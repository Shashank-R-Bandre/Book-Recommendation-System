# Popularity Based Recommender System

#### Merging ratings and books dataset

```python
ratings.merge(books, on='ISBN')
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.groupby('Book-Title').count()
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating']
```

```python
ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

#### Total number of ratings for each book

```python
number_of_ratings = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

#### Rename the columns

```python
number_of_ratings.rename(columns={'Book-Rating':'number_of_ratings'},inplace=True)
```

```python
number_of_ratings
```

#### Average ratings for each books

```python
ratings_with_name.groupby('Book-Title').mean()
```

```python
ratings_with_name.groupby('Book-Title').mean()['Book-Rating']
```

```python
ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_ratings = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_ratings.rename(columns={'Book-Rating':'average_ratings'},inplace=True)
```

```python
avg_ratings
```

```python
popular_books = number_of_ratings.merge(avg_ratings, on='Book-Title')
popular_books
```

#### Books with number of ratings >= 250 

```python
popular_books['number_of_ratings'] >= 250
```

```python
popular_books[popular_books['number_of_ratings'] >= 250]
```

#### Sorting in descending order on the basis of average_ratings

```python
popular_books[popular_books['number_of_ratings'] >= 250].sort_values('average_ratings',ascending=False)
```

```python
popular_books[popular_books['number_of_ratings'] >= 250].sort_values('average_ratings',ascending=False).head(100)
```

```python
popular_books['average_ratings'] = popular_books['average_ratings'].round(decimals = 2)
```

```python
popular_books.head()
```

```python
popular_books[popular_books['number_of_ratings'] >= 250].sort_values('average_ratings',ascending=False).head(100)
```

```python
top_100 = popular_books[popular_books['number_of_ratings'] >= 250].sort_values('average_ratings',ascending=False).head(100)
```

```python
top_100
```

```python
top_100.merge(books, on='Book-Title').drop_duplicates('Book-Title')
```

```python
top_100.merge(books, on='Book-Title').drop_duplicates('Book-Title').shape
```

```python
top_100.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'number_of_ratings', 'average_ratings', 'Image-URL-M']]
```

```python
data = top_100.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'number_of_ratings', 'average_ratings', 'Image-URL-M']]
```

```python
data.head(3)
```
